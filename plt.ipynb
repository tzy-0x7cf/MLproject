```python
import matplotlib.pyplot as plt

# Results for Project_dataset1
labels = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC']
project1_scores = [97.37, 98.27, 92.97, 96.39, 97.05]

# Results for Project_dataset2
project2_scores = [73.58, 65.26, 55.00, 58.30, 69.23]

# Best algorithms for each metric
best_algorithms_project1 = ['Logistic Regression', 'Logistic Regression', 'KNN', 'Logistic Regression', 'SVM']
best_algorithms_project2 = ['Logistic Regression', 'Logistic Regression', 'KNN', 'Logistic Regression', 'Logistic Regression']

width = 0.35  
x = range(len(labels))

fig, ax = plt.subplots(figsize=(10, 6))

rects1 = ax.bar(x, project1_scores, width, label='Project_dataset1', alpha=0.7)
rects2 = ax.bar([i + width for i in x], project2_scores, width, label='Project_dataset2', alpha=0.7)

for i, rect in enumerate(rects1):
    ax.annotate(best_algorithms_project1[i], xy=(rect.get_x() + rect.get_width() / 2, rect.get_height()),
                xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')

for i, rect in enumerate(rects2):
    ax.annotate(best_algorithms_project2[i], xy=(rect.get_x() + rect.get_width() / 2, rect.get_height()),
                xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')

ax.set_ylabel('Scores (%)')
ax.set_title('Best Algorithm Performance Comparison')
ax.set_xticks([i + width/2 for i in x])
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
